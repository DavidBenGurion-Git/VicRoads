```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(rc={'figure.figsize':(16.7,13.27)})
```

```python
lane2 = pd.read_csv("Lane2.csv")
lane2['Detector_Id'] = lane2['Detector_Id'].astype(str)
```

```python
lane2 = lane2.drop(columns = ['ID', 'Date','Occupancy','Speed_Sum','Speed_Obs','Configuration_Id','Available','Incident','Failed'])
```

```python
lane2.head()
```

```python
data = pd.DataFrame({'X': lane2['Detector_Id'], 'Y': lane2['Time'], 'Z': lane2['Volume']})
data_pivoted = data.pivot("X", "Y", "Z")

data_pivoted.head(20)
```

```python
ax = sns.heatmap(data_pivoted)
ax.set(xlabel='Time', ylabel='Detector ID', title='Lane2 Volume');
```
