```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(rc={'figure.figsize':(16.7,13.27)})
```

```python
lane4 = pd.read_csv("Lane4.csv")
lane4['Detector_Id'] = lane4['Detector_Id'].astype(str)
```

```python
lane4 = lane4.drop(columns = ['ID', 'Date','Occupancy','Speed_Sum','Speed_Obs','Configuration_Id','Available','Incident','Failed'])
```

```python
lane4.head()
```

```python
data = pd.DataFrame({'X': lane4['Detector_Id'], 'Y': lane4['Time'], 'Z': lane4['Volume']})
data_pivoted = data.pivot("X", "Y", "Z")

data_pivoted.head(20)
```

```python
ax = sns.heatmap(data_pivoted)
ax.set(xlabel='Time', ylabel='Detector ID', title='Lane4 Volume');
```
