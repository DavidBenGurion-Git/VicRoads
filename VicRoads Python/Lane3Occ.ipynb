```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(rc={'figure.figsize':(16.7,13.27)})
```

```python
lane3 = pd.read_csv("Lane3.csv")
lane3['Detector_Id'] = lane3['Detector_Id'].astype(str)
```

```python
lane3 = lane3.drop(columns = ['ID', 'Date','Volume','Speed_Sum','Speed_Obs','Configuration_Id','Available','Incident','Failed'])
```

```python
lane3.head()
```

```python
data = pd.DataFrame({'X': lane3['Detector_Id'], 'Y': lane3['Time'], 'Z': lane3['Occupancy']})
data_pivoted = data.pivot("X", "Y", "Z")
```

```python
ax = sns.heatmap(data_pivoted)
ax.set(xlabel='Time', ylabel='Detector ID', title='Lane3 Occupancy');
```
