```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(rc={'figure.figsize':(16.7,13.27)})
```

```python
lane5 = pd.read_csv("Lane5.csv")
lane5['Detector_Id'] = lane5['Detector_Id'].astype(str)
```

```python
lane5 = lane5.drop(columns = ['ID', 'Date','Volume','Speed_Sum','Speed_Obs','Configuration_Id','Available','Incident','Failed'])
```

```python
lane5.head()
```

```python
data = pd.DataFrame({'X': lane5['Detector_Id'], 'Y': lane5['Time'], 'Z': lane5['Occupancy']})
data_pivoted = data.pivot("X", "Y", "Z")
```

```python
data_pivoted.head(20)
```

```python
df2= data_pivoted.reindex(['1109704','1110858','1108471','1108464','1109515','1097096','1108480','1097144'])
```

```python
ax = sns.heatmap(df2)
ax.set(xlabel='Time', ylabel='Detector ID', title='Lane5 Occupancy');
```
